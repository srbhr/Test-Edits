```python
def get_machine_translations(m, n):
    from Data import Data
    dataset = Data().get_data()
    out = []
    for a in range(m, n):
        out.append(dataset[a][1])
    return out
```

```python
def get_post_editings(m, n):
    from Data import Data
    dataset = Data().get_data()
    out = []
    for a in range(m, n):
        out.append(dataset[a][2])
    return out
```

```python
mt = get_machine_translations(0,10)
```

```python
pe = get_post_editings(0, 10)
```

```python
def words_to_vec(x):
    from WordEmbeddings import WordEmbeddings
    out = []
    for a in x:
        out.append(WordEmbeddings(text=a).get_spanish_vectors(normalized=False))
    return out
```

```python
vectors_mt = words_to_vec(mt)
```
