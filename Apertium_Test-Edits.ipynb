```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
from WordEmbeddings import WordEmbeddings
```

```python
def get_translations(start=0, end=1, type='mt'):
    from Data import Data
    dataset = Data().get_data()
    out = []
    if type == 'mt':
        for a in range(start, end):
            out.append(dataset[a][1])
        return out
    else:
        for a in range(start, end):
            out.append(dataset[a][2])
        return out
    
def make_plot(mt_vectors, pt_vectors, color1='green', color2='yellow'):
    x = [i for i in range(len(mt_vectors))]
    x2 = [i for i in range(len(pt_vectors))]
    plt.figure(figsize=(16,8))
    sns.lineplot(x=x, y=mt_vectors, marker='o', linewidth=3.5, label="Machine Translated",color='green', markersize=10)
    sns.lineplot(x=x2, y=pt_vectors, marker='X',label="Post Edited",color='orange', markersize=12)
    
def get_plot_data(mt_text, pt_text):
    mt_vectors = WordEmbeddings(text=mt_text).get_spanish_vectors(remove_text=True)
    pt_vectors = WordEmbeddings(text=pt_text).get_spanish_vectors(remove_text=True)
    print("Macine Translated : ", mt_text)
    print('\n ')
    print("Post-Edited : ", pt_text, '\n \n')
    make_plot(mt_vectors, pt_vectors)
```

```python
mt = get_translations(start=0, end=10, type='mt')
pt = get_translations(start=0, end=10, type='pt')
```
